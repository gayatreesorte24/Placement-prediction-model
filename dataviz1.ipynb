# DATA VISUALIZATION OF PLACEMENT PREDICTION

# Importing libraries|

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Reading the dataset

```python
data = pd.read_csv('placed.csv')
```

```python
data.info()
```

# Distribution plot

```python
sns.distplot(data['ssc_p'])
```

```python
sns.distplot(data['hsc_p'])
```

```python
sns.distplot(data['degree_p'])
```

# Joint plot

```python
sns.jointplot(x='ssc_p',y='hsc_p',data=data,kind='scatter')
```

```python
sns.jointplot(x='hsc_p',y='degree_p',data=data,kind='hex')
```

# Pair plot

```python
sns.pairplot(data)
```

# Bar plot

```python
sns.barplot(x='ssc_b',y='ssc_p',data=data)
```

```python
sns.barplot(x='hsc_b',y='hsc_p',data=data)
```

# Count plot

```python
plt.figure(figsize=(9,5))
sns.countplot(x='skill1',data=data)
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x='skill3',data=data)
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x='skill4',data=data)
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x='status',data=data)
```

# Heat map

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.corr())
```
